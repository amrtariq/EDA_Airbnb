## Cleaning

as seen from data above coloumns 'calendar_updated', 'bathrooms', 'neighbourhood_group_cleansed' data are na so need to be dropped

```python
# Set display options to show all rows and columns 
pd.set_option('display.max_rows', 75)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)  # Adjust display width if needed
```

```python
df = df.drop(['calendar_updated', 'bathrooms', 'neighbourhood_group_cleansed'], axis = 1)
```

```python
pd.DataFrame(df.isna().sum())
```

remove small null values specially in dates

```python
checks = ['first_review','last_review', 'review_scores_cleanliness','reviews_per_month']
df = df.dropna(subset = checks)
```

```python
# Set display options to show all rows and columns 
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)  # Adjust display width if needed

pd.DataFrame(df.isna().sum())
```

Seperate Day/Month/Year

```python
df['host_since'] = pd.to_datetime(df['host_since'])
df['hsinceDay'] = df['host_since'].dt.day
df['hsinceMonth'] = df['host_since'].dt.month
df['hsinceYear'] = df['host_since'].dt.year

df['calendar_last_scraped'] = pd.to_datetime(df['calendar_last_scraped'])
df['clastScrapDay'] = df['calendar_last_scraped'].dt.day
df['clastScrapMonth'] = df['calendar_last_scraped'].dt.month
df['clastScrapYear'] = df['calendar_last_scraped'].dt.year

df['first_review'] = pd.to_datetime(df['first_review'])
df['fReviewDay'] = df['first_review'].dt.day
df['fReviewMonth'] = df['first_review'].dt.month
df['fReviewYear'] = df['first_review'].dt.year

df['last_review'] = pd.to_datetime(df['last_review'])
df['lReviewDay'] = df['last_review'].dt.day
df['lReviewMonth'] = df['last_review'].dt.month
df['lReviewYear'] = df['last_review'].dt.year
df.head(8)
```

```python
df['last_scraped'] = pd.to_datetime(df['last_scraped'])
df['lastScrapDay'] = df['last_scraped'].dt.day
df['lastScrapMonth'] = df['last_scraped'].dt.month
df['lastScrapYear'] = df['last_scraped'].dt.year
df = df.drop('last_scraped', axis = 1)
```

```python
# Fill missing values based on dtype
for col in df.columns:
    if df[col].dtype == 'object':
        mode_value = df[col].mode()[0]  # Mode for object columns
        df[col].fillna("NA", inplace=True)
    else:
        mean_value = df[col].mean()  # Mean for numeric columns
        df[col].fillna(mean_value, inplace=True)
```

```python
#Clean string extras
df['price'] = df['price'].str.replace('$', '').str.replace(',', '').astype(float)
bath_keywords= ['NA','Half-bath','half-bath','bathrooms_number',' baths',' bath', 'Private','private','shared', 'Shared',' ' ]
df['bathrooms_number'] = df['bathrooms_text']
for item in bath_keywords:
    df['bathrooms_number'] = df['bathrooms_number'].str.replace(item, '')
    
df['bathrooms_number'] = df['bathrooms_number'].replace('', -1)
df['bathrooms_number'] = df['bathrooms_number'].astype(float)

```
